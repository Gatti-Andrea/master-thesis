```python
# Show plots inline, and load main getdist plot module and samples class
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import sys, os
sys.path.insert(0,os.path.realpath(os.path.join(os.getcwd(),'..')))
from getdist import plots, MCSamples
import getdist
import matplotlib.pyplot as plt
import IPython
print('GetDist Version: %s, Matplotlib version: %s'%(getdist.__version__, plt.matplotlib.__version__))
```

```python
# Loading samples
chains_dir = os.path.realpath(os.getcwd())

samples = getdist.mcsamples.loadMCSamples(file_root=os.path.join(chains_dir, 'supernovae'), settings={'ignore_rows':0.3})
print('Parameters: ', samples.paramNames.names)
```

```python
# make the plots
g = plots.get_single_plotter()
g.plot_2d(samples, ['omegam', 'H0']) # put whatever parameters you want to plot
```

```python
# 1D marginalized plot
g = plots.get_single_plotter(width_inch=4)
g.plot_1d(samples, 'omegam') # put whatever parameter you want to plot
```

```python
g = plots.get_subplot_plotter()
g.triangle_plot(samples, ['w', 'wa', 'omegam', 'H0'], filled=True)
```
